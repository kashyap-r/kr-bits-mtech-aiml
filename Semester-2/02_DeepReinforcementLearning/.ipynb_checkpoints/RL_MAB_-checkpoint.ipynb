```python
import numpy as np

# Define the probabilities of reward for each arm
true_probabilities = [0.3, 0.5, 0.7]

# Function to simulate pulling an arm
def pull_arm(arm):
    return np.random.rand() < true_probabilities[arm]

# Greedy strategy function with transition logging
def greedy_strategy_with_logging(num_arms, num_pulls):
    estimated_values = np.zeros(num_arms)  # Estimated values of each arm
    num_times_pulled = np.zeros(num_arms)  # Number of times each arm has been pulled
    transitions = []  # Log of transitions (selected arm and reward)

    # Initial exploration: pull each arm once
    for arm in range(num_arms):
        reward = pull_arm(arm)
        estimated_values[arm] = reward
        num_times_pulled[arm] = 1
        transitions.append((arm, reward, estimated_values.copy(), num_times_pulled.copy()))

    # Continue pulling arms based on greedy strategy
    for _ in range(num_pulls - num_arms):
        # Select the arm with the highest estimated value
        best_arm = np.argmax(estimated_values)
        reward = pull_arm(best_arm)

        # Update the estimated value of the selected arm
        num_times_pulled[best_arm] += 1
        estimated_values[best_arm] += (reward - estimated_values[best_arm]) / num_times_pulled[best_arm]

        # Log the transition
        transitions.append((best_arm, reward, estimated_values.copy(), num_times_pulled.copy()))

    return estimated_values, num_times_pulled, transitions

# Run the updated greedy strategy with logging
num_arms = len(true_probabilities)
num_pulls = 1000
estimated_values, num_times_pulled, transitions = greedy_strategy_with_logging(num_arms, num_pulls)

# Display a few transitions for illustration
for i, (arm, reward, est_values, times_pulled) in enumerate(transitions[:500]):
    print(f"Round {i+1}: Selected Arm = {arm}, Reward = {reward}")
    print(f"Estimated Values: {est_values}")
    print(f"Number of Times Each Arm Was Pulled: {times_pulled}")
    print()

print("Final Estimated Values:", estimated_values)
print("Final Number of Times Each Arm Was Pulled:", num_times_pulled)
```
