```python
# read 
import os
from docx import Document

def read_docx(file_path):
    doc = Document(file_path)
    content = []
    for para in doc.paragraphs:
        content.append(para.text)
    return '\n'.join(content)

def read_multiple_docx(directory):
    documents_content = {}
    for filename in os.listdir(directory):
        if filename.endswith('.docx'):
            file_path = os.path.join(directory, filename)
            documents_content[filename] = read_docx(file_path)
    return documents_content

# Usage example
directory_path = 'path/to/your/word/documents'
documents = read_multiple_docx(directory_path)

for filename, content in documents.items():
    print(f"Content of {filename}:\n{content}\n")
```

```python
# will lead to losing any formatting information or any other metadata from the text, if extracting text is the only aim then this should work.

from docx2python import docx2python
doc_result = docx2python('page-wise-file.docx')
count = 0
para = 0
pages= []
while para < len(doc_result.body[0][0][0]):
    if doc_result.body[0][0][0][para] != "":
        current_page = {}
        current_page_paras = []
        count+=1
        while doc_result.body[0][0][0][para]!= "" and para<len(doc_result.body[0][0][0]):
            current_page_paras.append(doc_result.body[0][0][0][para])
            para+=1
        current_page["page_text"] = "\n".join(current_page_paras)
        current_page["page_no"] = count
        pages.append(current_page)
    else:
        para+=1
```
