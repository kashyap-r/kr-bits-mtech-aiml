```python
import spacy
import pandas as pd
import random
from spacy import displacy

# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Load the dataset
colnames = ['Sentiment', 'News Headline']
data = pd.read_csv("all-data.csv",  names=colnames, encoding="windows_1258")
df = data
# Filter sentences with at least 10 words
long_sentences = df[df['News Headline'].str.split().str.len() >= 10]['News Headline']

# Randomly select two sentences
random_sentences = random.sample(long_sentences.tolist(), 2)

# Parse and plot
for i, sentence in enumerate(random_sentences):
    doc = nlp(sentence)
    
    # Generate and save the HTML visualization
    html = displacy.render(doc, style='dep', page=True)
    with open(f'sentence_{i+1}_statistical_parse.html', 'w') as f:
        f.write(html)
        
    print(f'Sentence {i+1} statistical parse saved to sentence_{i+1}_statistical_parse.html')
```

```python
import spacy
import pandas as pd
import random
from spacy import displacy

# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Load your dataset
# df = pd.read_csv('financial_phrase_bank.csv')

# Filter sentences with at least 10 words
long_sentences = df[df['News Headline'].str.split().str.len() >= 10]['News Headline']

# Randomly select two sentences
random_sentences = random.sample(long_sentences.tolist(), 2)

# Parse and plot
for i, sentence in enumerate(random_sentences):
    doc = nlp(sentence)
    
    # Generate the HTML visualization
    html = displacy.render(doc, style='dep', page=True)
    
    # Check if html is not None
    if html:
        # Save HTML to file
        with open(f'sentence_{i+1}_statistical_parse.html', 'w') as f:
            f.write(html)
        print(f'Sentence {i+1} statistical parse saved to sentence_{i+1}_statistical_parse.html')
    else:
        print(f'Error generating HTML for Sentence {i+1}')
```

```python
from spacy import displacy

# Serve the visualization on a local server (for testing)
for i, sentence in enumerate(random_sentences):
    doc = nlp(sentence)
    displacy.serve(doc, style='dep')
    print(f'Sentence {i+1} displayed in a local server')
```
