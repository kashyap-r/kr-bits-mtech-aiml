```python
import stanza
import matplotlib.pyplot as plt
from stanza.utils.conll import CoNLL
from stanza.utils.visualize import DependencyTree

# Download English model
stanza.download('en')
nlp = stanza.Pipeline('en')

import pandas as pd
# Load the dataset
colnames = ['Sentiment', 'News Headline']
data = pd.read_csv("all-data.csv",  names=colnames, encoding="windows_1258")

# Filter sentences with at least 10 words
long_sentences = df[df['News Headline'].str.split().str.len() >= 10]['News Headline']

# Randomly select two sentences
import random
random_sentences = random.sample(long_sentences.tolist(), 2)

# Parse and plot
for i, sentence in enumerate(random_sentences):
    doc = nlp(sentence)
    
    # Convert to CoNLL format
    conll_data = CoNLL.convert_stanza(doc)
    
    # Create DependencyTree
    tree = DependencyTree(conll_data)
    
    # Plot dependency parse
    plt.figure(figsize=(12, 8))
    tree.plot()
    plt.title(f'Sentence {i+1}')
    plt.show()
```

```python
import nltk
from nltk import DependencyGraph
import matplotlib.pyplot as plt

# Sample dependency tree for illustration
def plot_nltk_dependency_tree():
    # Create a dependency graph
    dependency_tree = DependencyGraph("""
    1	ROOT	root	_	_	_	0	root	_	_
    2	am	am	_	_	_	1	root	_	_
    3	a	a	_	_	_	2	am	_	_
    4	sample	sample	_	_	_	2	am	_	_
    5	sentence	sentence	_	_	_	4	sample	_	_
    """)
    
    # Draw the tree
    plt.figure(figsize=(12, 8))
    dependency_tree.tree().pretty_print()
    plt.show()

plot_nltk_dependency_tree()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Sample data
edges = [("ROOT", "jumps"), ("jumps", "fox"), ("fox", "brown"), ("brown", "quick"), ("jumps", "over"), ("over", "dog")]

# Create a directed graph
G = nx.DiGraph()
G.add_edges_from(edges)

# Plot the graph
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=12, font_weight='bold', arrows=True)
plt.title('Dependency Parse Tree')
plt.show()
```
