## Part I
Sentence completion using N-gram: 

    Recommend the top 3 words to complete the given sentence using N-gram language model. 
    The goal is to demonstrate the relevance of recommended words based on the occurrence of Trigram within the corpus. Use all the instances in the dataset as a training corpus.

    Test Sentence: “ I love these cookies. I am on the paleo diet right now and these cookies --------“

#### N-Gram

    In Natural Language Processing (NLP), an n-gram is a contiguous sequence of n items (or words) from a given sample of text or speech. These items can be characters, syllables, words, or even symbols. The "n" in n-gram stands for the number of these items in the sequence.


```python
# Combine all text into a single corpus
corpus = ' '.join(df['Text'].astype(str))

# Tokenize the corpus
tokens = nltk.word_tokenize(corpus.lower())

# Create n-grams (up to 5-grams)
n_grams = [list(ngrams(tokens, i)) for i in range(1, 6)]

# Build n-gram models
n_gram_models = [defaultdict(Counter) for _ in range(5)]

for i, grams in enumerate(n_grams):
    for gram in grams:
        if i == 0:
            n_gram_models[i][()][gram] += 1
        else:
            n_gram_models[i][gram[:-1]][gram[-1]] += 1

# Function to predict next word
def predict_next_word(words, n=3):
    predictions = []
    context = tuple(words[-4:])  # Use up to 4 words of context

    for i in range(min(len(context), 4), -1, -1):
        current_context = context[-i:]
        if current_context in n_gram_models[i]:
            candidates = n_gram_models[i][current_context].most_common(10)
            for word, _ in candidates:
                if word not in predictions:
                    predictions.append(word)
                    if len(predictions) == n:
                        return predictions

    return predictions[:n]

# Test sentence
test_sentence = "I love these cookies. I am on the paleo diet right now and these cookies"
test_tokens = nltk.word_tokenize(test_sentence.lower())

# Predict next words
predicted_words = predict_next_word(test_tokens)

print("The input test sentence: \n", test_sentence)
print("\nTop 3 predicted words:", predicted_words)

```

## Part II 

i) Text Preprocessing 

a. Tokenization
b. Lowercasing 
c. Stop Words Removal
d. Stemming
e. Lemmatization

• Tokenization: Splitting the text into tokens (words) helps in handling each word individually for further processing.

• Stopword Removal: Removing common words (like “the”, “and”, etc.) that do not contribute significantly to the meaning of the text helps in
reducing noise.

• Stemming and Lemmatization:

    • Stemming: Reduces words to their root form, which can help in matching words with similar meanings but different forms (e.g., “running” to  “run”).
    • Lemmatization: Converts words to their base or dictionary form, ensuring that words are analyzed in their most meaningful form.

The below method performs the text-preprocessing tasks as listed above.

Note: The input data is converted to Lowercase in the below method before tokenization 

```python

# Function for comprehensive text preprocessing
def preprocess_text(text):
    # Tokenization
    tokens = nltk.word_tokenize(text.lower())
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    # Remove non-alphabetic characters
    tokens = [re.sub(r'[^a-zA-Z]', '', token) for token in tokens]
    tokens = [token for token in tokens if token]
    # Stemming
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return stemmed_tokens, lemmatized_tokens

# Apply preprocessing to each text in the dataset
df['preprocessed_stemmed'], df['preprocessed_lemmatized'] = zip(*df['Text'].apply(preprocess_text))
```

## Part II

ii) Feature Extraction 

    Use the pre-processed data from previous step and implement the below vectorization methods to extract features.
    **Word Embedding using Skip Gram **

Solution Approach on Feature Extraction:

1. Word Embeddings using Word2Vec:
    • Skip-Gram Model: The Skip-Gram model predicts context words from a given word, which is beneficial for capturing the semantic relationships
   between words. This method is well-suited for smaller datasets and can effectively capture the nuances of word meanings in the context of product
   reviews.

   • Word2Vec Benefits:
       • Semantic Richness: Word2Vec captures semantic and syntactic similarities between words. Words with similar meanings are placed close to each
       other in the vector space.
       • Dimensionality Reduction: By converting words into fixed-length vectors, Word2Vec facilitates the application of various machine learning
       algorithms that require numerical input.
   
2. Relevance Filtering:
    • Relevant Terms Set: By defining a set of relevant terms, the model can focus on words that are contextually important for the given task (e.g.,
   words related to cookies, baking, etc.). This ensures that the similarity measure prioritizes words that are meaningful in the context of the reviews.

```python
# Train Word2Vec model using Skip-Gram
model_sg = Word2Vec(df['preprocessed_lemmatized'].tolist(), vector_size=100, window=5, min_count=5, workers=4, sg=1, negative=10)

# Count word frequencies in the corpus
word_freq = Counter([word for text in df['preprocessed_lemmatized'] for word in text])
```

####   Solution Approach on Similarity Metric:
    
1. Cosine Similarity:
    • Why Cosine Similarity?
           Cosine similarity measures the cosine of the angle between two non-zero vectors, providing a similarity measure that is independent of their
   magnitude. This is particularly useful in text analysis where word vectors (embeddings) represent words in a high-dimensional space.

    • Benefits:
           • Magnitude Independence: It focuses on the orientation (or direction) of the vectors rather than their magnitude. This is important as two
           words can be similar even if their usage frequency is different.
    • Efficiency:
           • Cosine similarity is computationally efficient to calculate, making it suitable for comparing large sets of word vectors.

2. Frequency Factor:
    • Why Incorporate Frequency?: Adjusting cosine similarity with a frequency factor ensures that common words (which might be less informative) do not
   dominate the similarity metric.
    • Log Transformation: Using a logarithmic transformation of word frequency helps to dampen the effect of very frequent words. This balances the
   influence of word frequency without completely disregarding it, ensuring that rare but contextually significant words get more weight in the
   similarity measure.

```python
# Function to get word vector
def get_word_vector(word):
    try:
        return model_sg.wv[word]
    except KeyError:
        print(f"Warning: '{word}' not in vocabulary")
        return None

# Custom similarity function
def custom_similarity(word1, word2):
    try:
        cos_sim = model_sg.wv.similarity(word1, word2)
    except KeyError:
        return 0.0
    freq_factor = np.log(word_freq[word2] + 1)  # Log to dampen the effect of very frequent words
    return cos_sim * freq_factor
```

```python
# Function to find top N similar words with relevance filtering
def find_similar_words(word, n=2):
    word_vector = get_word_vector(word)
    if word_vector is None:
        return []

    # Define a set of relevant terms
    relevant_terms = {'cookie', 'bake', 'sweet', 'taste', 'flavor', 'chocolate', 'sugar', 'crunchy', 'soft', 'delicious', 'snack', 'dessert'}

    similarities = []
    for w in model_sg.wv.index_to_key:
        if w != word:
            sim = custom_similarity(word, w)
            if w in relevant_terms or w.endswith('y'):
                similarities.append((w, sim))

    return sorted(similarities, key=lambda x: x[1], reverse=True)[:n]

# Example: Get vector for a word
cookie_vector = get_word_vector('cookie')
if cookie_vector is not None:
    print("Vector for 'cookie':", cookie_vector[:5])  # Print first 5 elements
else:
    print("Could not get vector for 'cookie'")

# Example: Find similar words
similar_words = find_similar_words('cookie')
print("Words similar to 'cookie':", similar_words)
```

### Visualization

PCA for Dimensionality Reduction: Principal Component Analysis (PCA) is used to reduce the dimensionality of the word vectors from 100 dimensions to 2, making it possible to visualize the word embeddings.

• Benefits of PCA:
    • Visualization: Reduces complexity while retaining the most important variance in the data, enabling visual exploration of the relationships
    between words.
    
    • Interpretability: Helps in understanding the distribution and clustering of word vectors in a lower-dimensional space.

```python
# Visualize a subset of vector embeddings
def visualize_embeddings(words):
    # Filter words that are in the vocabulary
    valid_words = [word for word in words if word in model_sg.wv]

    if not valid_words:
        print("None of the specified words are in the vocabulary.")
        return

    vectors = [model_sg.wv[word] for word in valid_words]

    pca = PCA(n_components=2)
    result = pca.fit_transform(vectors)

    plt.figure(figsize=(10, 8))
    plt.scatter(result[:, 0], result[:, 1])
    for i, word in enumerate(valid_words):
        plt.annotate(word, xy=(result[i, 0], result[i, 1]))
    plt.title('Word Embeddings visualization')
    plt.show()

# Example: Visualize embeddings for a subset of words
subset_words = ['cookie', 'paleo', 'diet', 'gluten', 'free', 'taste']
visualize_embeddings(subset_words)

# Print some diagnostic information
print("Vocabulary size:", len(model_sg.wv.key_to_index))
print("Some common words:", list(model_sg.wv.key_to_index.keys())[:10])
print("Sample preprocessed text:", df['preprocessed_lemmatized'].iloc[0])
```
