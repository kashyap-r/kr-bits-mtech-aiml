```python
Elementary Row operations 
1. Interchange of two rows
2. Addition of a constant multiple of one row to another row
3. Multiplication of a row by a non-zero constant
```

```python
Row Echelon Form 
1. Any rows of all zeros are below any other non-zero rows 
2. Each leading entry of a row is in a column to the right 
of the leading entry of the row above it
3. All entries in a column below a leading entry are zeroes
```

```python
Reduced Row Echelon Form (RREF)
We say that a matrix is in Reduced Row Echelon Form if it is in Echelon form and additionally,
     1.  The leading entry in each row is 1.
     2.  Each leading 1 is the only non zero entry in its        
          column

Note: No matter what sequence of row operations we use each matrix is row equivalent to one and only one reducedrow echelon matrix

    
            
```

```python
Rank of a Matrix 

The number of non-zero rows, r, 
in a RREF or coefficient matrix R is called rank of R and also the rank of A

The Rank is Invariant under elementary row operations 
```
