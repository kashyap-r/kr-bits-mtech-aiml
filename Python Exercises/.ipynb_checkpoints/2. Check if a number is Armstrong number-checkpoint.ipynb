## Armstrong number is a number that is equal to the sum of cubes of its digits. 

Given a number, check whether the given number is an Armstrong number or not! 

A positive integer is called an Armstrong number of order n if:
    abcd... = an + bn + cn + dn + ...

Example: 153 = 1 * 1 * 1 + 5 * 5 * 5 + 3 * 3 * 3

153 is an Armstrong number of order 3.

Inputs from the user will be number and order n.

Armstrong number is a number that is equal to the sum of cubes of its digits. 

For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

```python
def isArmstrong(num):
    arm = 0
    while (num != 0):
        rem = int(num % 10)        
        arm += rem * rem * rem 
        # num = int(num/10)                
        num //= 10    
    return arm

if __name__ == "__main__":
    num = 153
    if num == isArmstrong(num):
        print (f"The number {num} is an armstrong number") 
    else:
        print (f"The number {num} is NOT an armstrong number")
```
