```python
# Given a number n, write a function to print all prime factors of n. 
# For example, if the input number is 12, then output should be “2 2 3”.

"""
Step 1: Find the (multiplicative) factors of a number 
Step 2: Identify and the print the prime numbers among them
"""

num = 6 

for i in range(1, num+1, 1):
    fact = num//i
    print (f"i={i}, fact={fact}")
    
    rem = fact % i
    if rem == 0:
        print (f"{i} is  prime factor")
```
