```python
## REF 
"""
Row Echelon Form
A matrix is in row echelon form (ref) when it satisfies the following conditions.

    The first non-zero element in each row, called the leading entry, is 1.
    Each leading entry is in a column to the right of the leading entry in the previous row.
    Rows with all zero elements, if any, are below rows having a non-zero element.
"""


m2  = [[0,3,4],[5,6,7],[8,0,-11]]
R = 3
C = 3 
# is it possible to find out the number of rows and columns of a given matrix 

for i in range(R):
    for j in range(C):
        print(m2[i][j], end = " ")
    print()
    
# check if a matrix is in row echelon form
# if the leading element is 0, then check if all the elements below it are zero 
# 
for i in range(R):
    for j in range(C):
        if i == 0 and i == j:
            if (m2[i][j] == 0):
                leadingElement = 0
                break
            else:
                # go to next row
                break
        else:
            if m2[i][j] != 0:
                m2[i][j] = m2[i][j] / m2[i][j]
                print (f"KRP (i,j) = ({i}, {j}) entity value is {m2[i][j]}")

for i in range(R):
    for j in range(C):
        print(m2[i][j], end = " ")
    print()
```
