```python
import numpy as np
x = np.array([[3, 4, 5]]).T 
print (f"x:\n{x}")

print (x[0][0])
print (x[1][0])
print (x[2][0])
rows, cols = x.shape
print (rows)
print (cols)

for i in range(rows):
    for j in range(cols):
        x[i][j] = pow(x[i][j], 2)
        
print (x)
```

```python
# -- multiplies two matrices 
u = numpy.array([[1],[1]])
A = numpy.array([[1,1],
[1,1]])
print (u)
print (A)
x = A@u
print (x)
```
